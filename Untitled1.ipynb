```python
data_dir = "Data"
images = []
labels = []

etudiants = os.listdir(data_dir)
etudiants.sort()
label_map = {name: idx for idx, name in enumerate(etudiants)}
print(label_map)
```

```python
#
for etudiant in etudiants:
    chemin = os.path.join(data_dir, etudiant)
    for fichier in os.listdir(chemin):
        img_path = os.path.join(chemin, fichier)
        try:
            # DeepFace extract
            embedding = DeepFace.represent(img_path, detector_backend='retinaface')
            # embedding est une liste d'un dict, on prend la valeur du vecteur
            encodage = embedding[0]["embedding"]
            images.append(encodage)
            labels.append(etudiant)
        except Exception as e:
            print(f"‚ùå Probl√®me avec {img_path} : {e}")

print(f"‚úÖ Total encodages valides : {len(images)}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import save_model, load_model
```

```python
# Exemple simple
mlp = Sequential([
    Dense(128, activation='relu', input_shape=(len(images[0]),)),
    Dense(len(set(labels)), activation='softmax')
])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y = encoder.fit_transform(labels)  # ‚Üê c'est ce 'y'
```

```python
import numpy as np

X = np.array(images, dtype=np.float32)  # transforme la liste de listes en array 2D
y = np.array(y, dtype=np.int32)         # assure que les labels sont bien un array d'entiers
```

```python
# Puis entra√Ænement
mlp.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
mlp.fit(X, y, epochs=20, batch_size=8)
```
